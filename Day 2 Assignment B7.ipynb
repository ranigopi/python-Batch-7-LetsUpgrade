```
 
 
print("DEFAULT FUNCTION OF LIST IS")
fruits= ["apple","mango","banana","orange","grapes"]
print ("list of fruits:", fruits)
print("reverse is ",fruits[::-1])
fruits.append("kiwi")
print("adding of list ", fruits)
fruits.extend(["watermelon","book"])
print("extend of list is ", fruits)
fruits.insert(1,"strawberry")
print ("insert element at index 1 ", fruits)
fruits.remove("book")
print ("removing book ", fruits)
print ("popping the list ", fruits.pop())
print("popping of 1st index ", fruits.pop(1))
fruits.sort()
print ("sorting the list ", fruits)
fruits.reverse()
print("reverse all ", fruits)
b=fruits.copy()
print ("copy of fruits ",b)
print ("length ",len(fruits))
print ("maximum is ",max(fruits))
print ("minimum is ",min(fruits))
print("index of apple is ",fruits.index("apple"))
print("count is ", fruits.count("apple"))
fruits.clear()
print ("clear list ", fruits)
print (list(range(10)))
```

```
print("DEFAULT FUNCTION OF STRING IS")
a="python programming"
print("string is :", a)
print("capitalise is",a.capitalize())
print ("slicing is ",a[0:5])
b="T"+a[1:]
print("addition of alphabet ",b)
print("uppercase ",a.upper())
print("find of a",a.find("a"))
print("lowercase is ",a.casefold())
print("count of p ",a.count("p"))
print("endwith s ",a.endswith("s"))
print("find on ",a.find("on"))
print("is there is an alpha",a.isalpha())
print("is there is an number",a.isnumeric())
print("start with t",a.startswith("t"))
print("index ",a.index("g"))
print("lower ",a.islower())
print ("upper ",a.isupper())
print ("space",a.isspace())
```

```
print("DEFAULT FUNCTION OF DICTIONARY")
dict = {'Name': 'abc', 'Age': 18, 'Class': '12'}
print ("dict['Name']: ", dict['Name'])
print ("dict['Age']: ", dict['Age'])
dict['Age'] = 8; 
dict['School'] = "DEF School"
print ("dict['Age']: ", dict['Age'])
print ("dict['School']: ", dict['School'])
del dict['Name'] 
print (dict)
dict.clear()  
print (dict)
del dict
dict1 = {"name":"pqr" ,"age":"17"} 
print("new dictionary is ",dict1) 
B=dict1.copy()
print("copy of dict",B)
seq = ('name', 'age', 'sex')
dict = dict.fromkeys(seq)
print ("New Dictionary : %s" %  str(dict))
dict = dict.fromkeys(seq, 10)
print ("New Dictionary : %s" %  str(dict))
dict = {'Name': 'Zara', 'Age': 27}
 
print ("Value : %s" %  dict.get('Age'))
print ("Value : %s" %  dict.get('Sex', "NA"))
print ("Value : %s" %  dict.items())
print ("Value : %s" %  dict.keys())
dict2 = {'Sex': 'female' }
 
dict.update(dict2)
print ("updated dict : ", dict)
dict = {'Sex': 'female', 'Age': 7, 'Name': 'Zara'}
print ("Values : ",  list(dict.values()))
```
