```
 
print("Assignment No 3:")
for item in range(1042000 , 70264826):
    sum = 0
    temp = item
    length = len(str(item))
    
    while temp > 0:
        digit = temp % 10
        sum += digit ** length
        temp //= 10
    if item == sum:
        print(item,"is an armstrong")
        break
```
